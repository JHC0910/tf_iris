```python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split 
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
iris = pd.read_excel("iris.xlsx")
```

```python
iris[:5]
```

```python
x = iris[["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]]
```

```python
x[:5]
```

```python
y = iris[["target"]]
```

```python
y[5:]
```

```python
iris_array = iris.values
x_array = x.values
labels = iris_array[:,5]
x = x_array[:]
```

```python
x[:5]
```

```python
type(x)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train[:10]
```

```python
x_test[:5]
```

```python
 y_train[:5]
```

```python
y_train2 = tf.keras.utils.to_categorical(y_train, num_classes = 3)
y_test2 = tf.keras.utils.to_categorical(y_test, num_classes = 3)
```

```python
y_train2[:5]
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(input_dim = 4, units = 100, activation=tf.nn.relu))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(units = 200, activation=tf.nn.relu))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(units = 200, activation=tf.nn.relu))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(units = 3, activation=tf.nn.softmax))
```

```python
model.compile(optimizer = 'adam', loss = tf.keras.losses.categorical_crossentropy, metrics = ['accuracy'])
```

```python
history = model.fit(x_train, y_train2,
          epochs=200,
          batch_size=128)
```

```python
score = model.evaluate(x_test, y_test2, batch_size=128)
```
